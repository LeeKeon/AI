```python
# -*- coding: utf-8 -*-

import os
os.environ['CUDA_VISIBLE_DEVICES'] = "2"

import tensorflow as tf
import numpy as np
import time

from cnn_model_test import Model
from data_loader import movie_data

def initialize_session():
    config = tf.ConfigProto()
    config.gpu_options.per_process_gpu_memory_fraction = 0.4
    return tf.Session(config=config)
```

```python
data = movie_data()
```

```python
train, label = data.get_train_dataset()

print(len(train))
```

```python
print(train[0].shape)
print(label[0].shape)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu',
                 input_shape=(134, 91, 3)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))
```

```python
#model.compile(loss='binary_crossentropy',
#              optimizer=keras.optimizers.Adam(),
#              metrics=['accuracy'])

model.compile(loss='categorical_crossentropy',
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])

model.summary()
```

```python
n = 5000
#model.fit(np.array(train[: n]), np.array(label[: n]), batch_size=64, epochs=20,
#          verbose=1, validation_split=0.1)

model.fit(np.array(train), np.array(label), batch_size=64, epochs=20,
          verbose=1, validation_split=0.1)
```

```python
test_x, test_y = data.get_test_dataset()
print(len(test_x))
print(len(test_y))
```

```python
pred = model.predict(np.array(test_x))
```

```python
pred_argmax = pred.argmax(axis=1)
test_y_argmax = test_y.argmax(axis=1)

cnt = 0
for r, p in zip(test_y_argmax, pred_argmax):
    print('real', data.labelid_2_genre[r], ', pred', data.labelid_2_genre[p])
    if(r==p):
        cnt += 1

print('Test Accuracy ', cnt/len(pred))
```
