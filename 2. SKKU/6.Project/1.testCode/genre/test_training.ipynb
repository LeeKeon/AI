```python
from data_loader import movie_data
import matplotlib.pyplot as plt
data = movie_data()
```

```python
import tensorflow as tf

class_size = 26
initializer = tf.random_uniform_initializer(-0.1, 0.1)

# Placeholder (X_Image,Y_Label)
X_Image = tf.placeholder(tf.float32, shape=[None, 134, 91, 3], name='X_Image')
Y_Lable = tf.placeholder(tf.int32, shape=[None, ], name='Y_Label')

# CNN Layer
conv_f1 = tf.get_variable(name='conv_f1', shape=[3, 3, 3, 128], dtype=tf.float32, initializer=initializer)
conv_b1 = tf.get_variable(name='conv_b1', shape=[128], dtype=tf.float32, initializer=initializer)
conv_l1 = tf.nn.leaky_relu(tf.nn.conv2d(X_Image, filter=conv_f1, strides=[1, 1, 1, 1], padding='SAME') + conv_b1)
conv_l1 = tf.nn.max_pool(conv_l1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
print(conv_l1)

conv_f2 = tf.get_variable(name='conv_f2', shape=[3, 3, 128, 256], dtype=tf.float32, initializer=initializer)
conv_b2 = tf.get_variable(name='conv_b2', shape=[256], dtype=tf.float32, initializer=initializer)
conv_l2 = tf.nn.leaky_relu(tf.nn.conv2d(conv_l1, filter=conv_f2, strides=[1, 1, 1, 1], padding='SAME') + conv_b2)
conv_l2 = tf.nn.max_pool(conv_l2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
print(conv_l2)

conv_f3 = tf.get_variable(name='conv_f3', shape=[3, 3, 256, 512], dtype=tf.float32, initializer=initializer)
conv_b3 = tf.get_variable(name='conv_b3', shape=[512], dtype=tf.float32, initializer=initializer)
conv_l3 = tf.nn.leaky_relu(tf.nn.conv2d(conv_l2, filter=conv_f3, strides=[1, 1, 1, 1], padding='SAME') + conv_b3)
conv_l3 = tf.nn.max_pool(conv_l3, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
print(conv_l3)

conv_f4 = tf.get_variable(name='conv_f4', shape=[3, 3, 512, 512], dtype=tf.float32, initializer=initializer)
conv_b4 = tf.get_variable(name='conv_b4', shape=[512], dtype=tf.float32, initializer=initializer)
conv_l4 = tf.nn.leaky_relu(tf.nn.conv2d(conv_l3, filter=conv_f4, strides=[1, 1, 1, 1], padding='SAME') + conv_b4)
conv_l4 = tf.nn.max_pool(conv_l4, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
print(conv_l4)

conv_l4 = tf.reshape(conv_l4, (-1, 9 * 6 * 512))
print(conv_l4)

# FC Layer
fc_l1 = tf.contrib.layers.fully_connected(conv_l4, 512, activation_fn=tf.identity)
hypothesis = tf.contrib.layers.fully_connected(fc_l1, class_size, activation_fn=tf.nn.softmax)
print(hypothesis)

# argmax output check
Y_Predict = tf.argmax(hypothesis, 1)

# loss with prob (CE)
# optimizer
Y_Lable_onehot = tf.one_hot(Y_Lable,class_size,1.0,0.0)
cost = tf.reduce_mean(-tf.reduce_sum(Y_Lable_onehot * tf.log(tf.clip_by_value(hypothesis, 1e-10, 1.0)), [1]))
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)

#Evaluation
is_correct = tf.equal(tf.argmax(hypothesis, 1), tf.argmax(Y_Lable_onehot, 1))
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

# session run
sess = tf.Session()
sess.run(tf.global_variables_initializer())

print('Start learning..')
for epoch in range(15):
    total_loss = 0
    for i in range(50):
        batch_xs, batch_ys = data.get_train()
        loss, _, hypothesis_, Y_Lable_onehot_ = sess.run([cost, optimizer, hypothesis, Y_Lable_onehot], feed_dict={X_Image:batch_xs, Y_Lable:batch_ys})
        total_loss += loss
        #print('pred', hypothesis_)
        #print('real', Y_Lable_onehot_)
        #print('loss', loss)
        #print('accuracy ', sess.run(accuracy, feed_dict={X_Image: batch_xs, Y_Lable: batch_ys}))

    batch_tx, batch_ty = data.get_test()
    print(epoch+1, ' total_loss ', loss / 50, ' test accuracy ', sess.run(accuracy, feed_dict={X_Image: batch_tx, Y_Lable: batch_ty}))
```
