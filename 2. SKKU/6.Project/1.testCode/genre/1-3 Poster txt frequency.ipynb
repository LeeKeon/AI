```python
import os
import re
import nltk
from nltk.corpus import stopwords


path = './dataset/poster_txt/War/'
f_list = os.listdir(path)
```

```python
cnt = {}
lines = []
for file in f_list:    
    with open(path + file, "r", encoding="ISO-8859-1") as fin:
        lines.append(fin.readline())
```

```python
stopwords = nltk.corpus.stopwords.words('english')

for line in lines:
    for word in line.split():
        word = word.lower() 
        if word.lower() not in stopwords and re.search('[A-Za-z0-9]$', word):
            word += ' War'
            if word in cnt:
                cnt[word] += 1
            else:
                cnt[word] = 1
```
