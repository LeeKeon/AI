```python
def content_stopwords(text):
    stopwords = nltk.corpus.stopwords.words('english')
    result = []
    
    for word in text:
        if word.lower() not in stopwords:
            result.append(word.lower())
            
    return result
```

```python
lala_ = content_stopwords(lala_tokens)
raiders_ = content_stopwords(raiders_tokens)

print(len(lala_tokens),len(lala_))
print(len(raiders_tokens),len(raiders_))
```

```python
lala_fdisk = nltk.FreqDist(lala_)
raiders_fdisk = nltk.FreqDist(raiders_)
```

```python
lala_train = lala_fdisk.most_common(2000)
raiders_train = raiders_fdisk.most_common(2000)
```
