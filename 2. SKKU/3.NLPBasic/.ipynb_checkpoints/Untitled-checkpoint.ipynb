```python
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
import nltk

myst_words = stopwords.words('english')
mylist = ['-',',','.','"','\'',';']
myst_words += mylist
myst_words
```

```python
def content_fraction(text, stopwords):
    #stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return content
```

```python
wordlists = PlaintextCorpusReader('.','avengers.txt')

avengers_words = content_fraction(set(wordlists.words()),myst_words)
avengers_words
#avengers_words = set(wordlists.words())
```

```python
pos = 0.0
neg = 0.0
obj = 0.0
cnt = 0

for word in avengers_words:
    aset = wn.synsets(word)
    for i in aset:
        cnt += 1
        senti = swn.senti_synset(i.name())
        pos += senti.pos_score()
        neg += senti.neg_score()
        obj += senti.obj_score()
        
print('Positive ', round(pos/(cnt-obj) * 100,2), '%')
print('Negative ', round(neg/(cnt-obj) * 100,2), '%')
print('Object ',  round(obj/cnt * 100,2), '%')
```
